```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('climate_1991_2016_USA.csv')
```

```python
print(data.columns)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(data['Temperature - (Celsius)'])
plt.title('USA Average Monthly Temperatures')
plt.ylabel('Celsius')

plt.subplot(1,2,2)
plt.plot(data['Rainfall - (MM)'])
plt.title('USA Average Monthly Rainfall')
plt.ylabel('Inches')

plt.show()
```

These graphs above are difficult to interpret due to the orientation of the data. There is no clear way to label the X-axis as labeling every month would result in over 300 labeled data points, and labeling every year does not accurately reflect how temperatures/rainfall are changing on an annual scale. 

```python
#create new DataFrame with monthly averages to yearly averages

yearly_data = pd.DataFrame()
year = []
yearly_avg_temp = []
yearly_avg_rain = []

for i in range(int((len(data['Temperature - (Celsius)'])/12))):
    year.append(1991 + i)
    yearly_avg_temp.append((sum(data['Temperature - (Celsius)'][i*12:(i*12)+12]))/12)
    yearly_avg_rain.append((sum(data['Rainfall - (MM)'][i*12:(i*12)+12]))/12)
    
yearly_data['Yearly Temperature'] = yearly_avg_temp
yearly_data['Yearly Rainfall'] = yearly_avg_rain
yearly_data['Year'] = year
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(yearly_data['Year'], yearly_data['Yearly Temperature']) 
plt.title('USA Average Annual Temperatures')
plt.ylabel('Celsius')

plt.subplot(1,2,2)
plt.plot(yearly_data['Year'], yearly_data['Yearly Rainfall']) 
plt.title('USA Average Annual Rainfall')
plt.ylabel('Inches')

plt.show()
```

Calculating annual changes by averagaing monthly temperatures and rainfall within each year allows for clearer representations of the data. It appears that a trend towards higher temperatures is observed between 1990 to 2016, however further statistical analyses would be needed to confirm these findings.

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(data['Temperature - (Celsius)'])
plt.title('USA Average Monthly Temperatures')
plt.xlabel('Celsius')

plt.subplot(1,2,2)
plt.hist(data['Rainfall - (MM)'])
plt.title('USA Average Monthly Rainfall')
plt.xlabel('Inches')

plt.show()
```

Histograms appear to show a non-normal distribution for the average monthly temperatures and a normal distribution for average monthly rainfall. Average monthly temperatures appear to exhibit a multi-modal distribution where temperatures aggregate in low, moderate, and high temperature groupings. 

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(yearly_data['Yearly Temperature']) 
plt.title('USA Average Annual Temperatures')
plt.xlabel('Celsius')

plt.subplot(1,2,2)
plt.hist(yearly_data['Yearly Rainfall']) 
plt.title('USA Average Annual Rainfall')
plt.xlabel('Inches')

plt.show()
```

Both annual temperature and rainfall appear to be right skewed, where a majority of the data is distributed around a leftward mean.
